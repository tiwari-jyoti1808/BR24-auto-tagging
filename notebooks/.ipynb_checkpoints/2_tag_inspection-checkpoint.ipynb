# Tag Separation into In-Text and Out-of-Text tags

```python
from tqdm import tqdm
tqdm.pandas()
```

```python
df.head(1)
```

```python
def separate_in_text_tags(row):
    tags_in_text = []
    for tag in row.tags:
        if tag in row.clean_text:
            tags_in_text.append(tag)
    return tags_in_text
```

```python
df['in_text'] = df.progress_apply(separate_in_text_tags, axis=1)
df['out_of_text'] = df.progress_apply(lambda row: [word for word in row.tags if word not in row.in_text], axis=1)
```

```python
df.head(1)
```

```python
df['in_text_percent'] = df.apply(lambda row: int((len(row.in_text)/len(row.tags))*100), axis=1)
df['out_of_text_percent'] = 100-df['in_text_percent']
```

# Parts-of-Speech of in_text tags

```python
pos_tags_all_in_text = []
all_in_text = df['in_text'].apply(pd.Series).stack().str.strip()

for tag in list(set(all_in_text)):
    t = nlp(tag)
    pos_tags_all_in_text.append(t[0].pos_)

c = dict(Counter(pos_tags_all_in_text))
c = dict(sorted(c.items(), key=lambda x: x[1], reverse=True))
c
```

```python
def check_common(a):
    print('With bayern_tags ', len(list(set(a) & set(bayern_tags))))
    print('With kultur_tags ', len(list(set(a) & set(kultur_tags))))
    print('With sport_tags ', len(list(set(a) & set(sport_tags))))
    print('With deutschland_welt_tags ', len(list(set(a) & set(deutschland_welt_tags))))
    print('With wirtschaft_tags ', len(list(set(a) & set(wirtschaft_tags))))
    print('With netzwelt_tags ', len(list(set(a) & set(netzwelt_tags))))
    print('With wissen_tags ', len(list(set(a) & set(wissen_tags))))
    print('With das_wichtigste_tags ', len(list(set(a) & set(das_wichtigste_tags))))
```
